```python
from scipy.sparse import hstack
```

```python
def add_time_feature(df, X_sparce):
    hour = df.apply(lambda ts:ts.hour)
    morning = ((hour>=7)&(hour<=11)).astype('int')
    day = ((hour>=12)&(hour<=18)).astype('int')
    evening = ((hour>=19)&(hour<=23)).astype('int')
    night = ((hour>=0)&(hour<=6)).astype('int')
    X = hstack([X_sparce, morning.values.reshape(-1,1),day.values.reshape(-1,1),
               evening.values.reshape(-1,1),night.values.reshape(-1,1)])
    return X
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
CV = CountVectorizer()
with open('train_sessions_txt.txt') as inp_train_file:
    X_train = CV.fit_transform(inp_train_file)
with open('test_sessions_txt.txt') as inp_test_file:
    X_test = CV.transform(inp_test_file)
print(X_train.shape, X_test.shape)
```

```python
y_train = train_data['target']
```

```python
time_test = test_data[['time1', 'time2','time3','time4','time5','time6','time7','time8','time9','time10']]
for i in range(time_test.shape[1]):
    time_test.iloc[:,i] = pd.to_datetime(time_test.iloc[:,i])
time_test
```

```python
X_train_new = add_time_feature(time['time1'].fillna(0), X_train)
X_test_new = add_time_feature(time_test['time1'].fillna(0), X_test)
```

```python
X_train_new.shape, X_test_new.shape
```

```python
time_test['delta'] = time_test.max(axis = 1) - time_test.min(axis = 1)
time_test['delta']
```

```python
time_test['delta']=time_test['delta'].dt.total_seconds()
time['delta']=time['delta'].dt.total_seconds()
```

```python
X_train_new = hstack([X_train_new, time['delta'].values.reshape(-1,1)])
X_test_new = hstack([X_test_new, time_test['delta'].values.reshape(-1,1)])
```

```python
X_train_new.shape, X_test_new.shape
```

```python
from sklearn.model_selection import cross_validate
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(random_state=17, C =1)
```

```python
cv_results = cross_validate(model,X_train_new, y_train, cv=5, scoring = 'roc_auc')
```

```python
cv_results['test_score'].mean()
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
c_value = np.logspace(-2, 2,10)
```

```python
random_search = RandomizedSearchCV(estimator = model, param_distributions={'C':c_value}, scoring = 'roc_auc',
                           n_jobs=-1, cv=5, verbose=1)
```

```python
random_search.fit(X_train_new, y_train)
```
